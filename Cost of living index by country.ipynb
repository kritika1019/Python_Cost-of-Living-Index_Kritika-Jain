```python
index_state = df.groupby(['Country'], as_index=False)['Cost of Living Index'].sum().sort_values(by='Cost of Living Index', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = index_state, x = 'Country',y= 'Cost of Living Index')
```

```python
index_state = df.groupby(['Country'], as_index=False)['Rent Index'].sum().sort_values(by='Rent Index', ascending=False).tail(10)

sns.set(rc={'figure.figsize':(18,3)})
sns.barplot(data = index_state, x = 'Country',y= 'Rent Index')
```

```python
index_state = df.groupby(['Country'], as_index=False)['Groceries Index'].sum().sort_values(by='Groceries Index', ascending=False).tail(10)

sns.set(rc={'figure.figsize':(18,3)})
sns.barplot(data = index_state, x = 'Country',y= 'Groceries Index')
```

```python
ind_melted = pd.melt(ind, id_vars=['Country'], value_vars=['Cost of Living Index', 'Rent Index', 'Cost of Living Plus Rent Index','Groceries Index', 'Restaurant Price Index', 'Local Purchasing Power Index'], var_name='Index Type', value_name='Value')

# Create the bar plot
plt.figure(figsize=(15, 6))
ax = sns.barplot(x='Index Type', y='Value', data=ind_melted, palette='viridis')

# Add data labels
for bars in ax.containers:
    ax.bar_label(bars)

plt.title('Cost of Living Index and Rent Index in India')
plt.xlabel('Index Type')
plt.ylabel('Value')
plt.show()
```
